```python
import os
import csv
import numpy as np
```

```python
# two times
# 5 websites
# 8 teams

website_dict = [[[[] for k in range(2)] for j in range(5)] for i in range(8)]
ping_dict = []
for i in range(8):
    website_dict.append({})
    ping_dict.append({})
for c, team in enumerate(os.listdir("Project3CommunityData")):
    print(team)
    for f in os.listdir(f'Project3CommunityData/{team}/clean_data/seq2'):
        if 'api_seq2.txt' in f:
            key = f[:-13]
            if key not in website_dict[c]:
                website_dict[c][key] = []
            with open(f'Project3CommunityData/{team}/clean_data/seq2/{f}', 'r') as ping:
                pingreader = csv.reader(ping)
                for row in pingreader:
                    website_dict[c][key].append(float(row[-1]))
        if 'ping_seq2.txt' in f:
            key = f[:-14]
            if key not in ping_dict[c]:
                ping_dict[c][key] = []
            with open(f'Project3CommunityData/{team}/clean_data/seq2/{f}', 'r') as ping:
                pingreader = csv.reader(ping)
                for row in pingreader:
                    try:
                        ping_dict[c][key].append(float(row[-1]))
                    except:
                        pass
```

```python
for i in range(8):
    print(f'Team {i+1}')
    for key in website_dict[i]:
        print(key)
        api_response_time = np.mean(website_dict[i][key])
        ping_response_time = np.mean(ping_dict[i][key])
        print(ping_response_time/api_response_time)
```
